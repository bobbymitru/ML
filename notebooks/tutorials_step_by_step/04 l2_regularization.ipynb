# Libraries

```python
# Data transformation
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
```

# Data

```python
x = np.arange(0, 10.1, 0.1)
y = np.array([a ** 3 - 10 * a ** 2 + 3 * a + 500 for a in x])
```

# Implementation

```python
N = 13
lmbd = 30
```

```python
X = np.array([[a ** n for n in range(N)] for a in x])
lmbd_I = np.array([[lmbd if i == j else 0 for j in range(N)] for i in range(N)])
lmbd_I[0][0] = 0
```

```python
X_train = X[::2]
Y_train = y[::2]
```

```python
A = np.linalg.inv(X_train.T @ X_train + lmbd_I)
w = Y_train @ X_train @ A
```
